## Section 1: Setup and Database Connection

Set up SQLite database and necessary imports for SQL operations.

```python
import sqlite3                       # Python module for SQLite database
import pandas as pd                 # For displaying SQL query results nicely
from datetime import datetime, date # For working with dates and timestamps

# Create an in-memory SQLite database (exists only during this session)
# For persistent database, use: sqlite3.connect('database_name.db')
conn = sqlite3.connect(':memory:')

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

print("✓ SQLite database connection established!")
print(f"Database type: SQLite (In-Memory)")
print(f"Connection status: {'Active' if conn else 'Inactive'}")
print("\nNote: We're using an in-memory database for this tutorial.")
print("All data will be lost when the notebook closes.")
```

## Section 2: Creating Tables

Learn how to create tables with different data types and constraints.

```python
# Create Table 1: Employees Table
create_employees = """
CREATE TABLE Employees (
    EmployeeID INTEGER PRIMARY KEY AUTOINCREMENT,  -- Unique identifier (primary key)
    FirstName TEXT NOT NULL,                        -- Text field, cannot be NULL
    LastName TEXT NOT NULL,                         -- Text field, cannot be NULL
    Email TEXT UNIQUE,                              -- Text field, must be unique
    Department TEXT,                                -- Department name
    Salary DECIMAL(10, 2),                          -- Decimal with 2 decimal places
    HireDate DATE,                                  -- Date field
    IsActive BOOLEAN DEFAULT 1                      -- Boolean (1=true, 0=false), default is true
)
"""

cursor.execute(create_employees)
conn.commit()
print("✓ Employees table created successfully!")
print("\nTable Structure:")
print("- EmployeeID: Integer (Primary Key)")
print("- FirstName: Text")
print("- LastName: Text")
print("- Email: Text (Unique)")
print("- Department: Text")
print("- Salary: Decimal")
print("- HireDate: Date")
print("- IsActive: Boolean")
print("\n" + "="*60 + "\n")

# Create Table 2: Departments Table
create_departments = """
CREATE TABLE Departments (
    DepartmentID INTEGER PRIMARY KEY AUTOINCREMENT,
    DepartmentName TEXT NOT NULL UNIQUE,
    Location TEXT,
    Manager TEXT
)
"""

cursor.execute(create_departments)
conn.commit()
print("✓ Departments table created successfully!")
print("\n" + "="*60 + "\n")

# Create Table 3: Projects Table
create_projects = """
CREATE TABLE Projects (
    ProjectID INTEGER PRIMARY KEY AUTOINCREMENT,
    ProjectName TEXT NOT NULL,
    Department TEXT,
    Budget DECIMAL(15, 2),
    StartDate DATE,
    EndDate DATE,
    Status TEXT DEFAULT 'Active'
)
"""

cursor.execute(create_projects)
conn.commit()
print("✓ Projects table created successfully!")
print("\n" + "="*60 + "\n")
```

## Section 4: Basic SELECT Queries

Learn how to retrieve data from tables using SELECT statements.

```python
# Helper function to display SQL query results
def display_sql_result(query, title):
    """Function to execute and display SQL query results as a formatted table"""
    result = pd.read_sql(query, conn)
    print(f"\n{title}")
    print(result.to_string(index=False))
    return result

# Query 1: Select all columns from a table
query1 = "SELECT * FROM Employees"
display_sql_result(query1, "1. SELECT ALL Columns from Employees Table")
print("\n" + "="*60 + "\n")

# Query 2: Select specific columns
query2 = "SELECT FirstName, LastName, Department, Salary FROM Employees"
display_sql_result(query2, "2. SELECT Specific Columns (FirstName, LastName, Department, Salary)")
print("\n" + "="*60 + "\n")

# Query 3: Select with alias (rename columns in output)
query3 = """
SELECT 
    FirstName AS 'First Name',
    LastName AS 'Last Name',
    Department AS 'Dept',
    Salary AS 'Annual Salary'
FROM Employees
"""
display_sql_result(query3, "3. SELECT with Column ALIAS (Rename columns)")
print("\n" + "="*60 + "\n")

# Query 4: DISTINCT - Get unique values
query4 = "SELECT DISTINCT Department FROM Employees"
display_sql_result(query4, "4. SELECT DISTINCT Departments (Remove duplicates)")
print("\n" + "="*60 + "\n")

# Query 5: Count rows
query5 = "SELECT COUNT(*) as 'Total Employees' FROM Employees"
display_sql_result(query5, "5. COUNT - Total number of employees")
print("\n" + "="*60 + "\n")

# Query 6: Limit results
query6 = "SELECT * FROM Employees LIMIT 3"
display_sql_result(query6, "6. LIMIT - Show only first 3 rows")
print("\n" + "="*60 + "\n")
```

## Section 5: WHERE Clauses and Filtering

Filter data using WHERE clause with various conditions.

```python
# Filter 1: Equal to condition
query_where1 = """
SELECT FirstName, LastName, Department, Salary
FROM Employees
WHERE Department = 'IT'
"""
display_sql_result(query_where1, "1. WHERE Department = 'IT' (Equal condition)")
print("\n" + "="*60 + "\n")

# Filter 2: Greater than condition
query_where2 = """
SELECT FirstName, LastName, Salary
FROM Employees
WHERE Salary > 65000
"""
display_sql_result(query_where2, "2. WHERE Salary > 65000 (Greater than condition)")
print("\n" + "="*60 + "\n")

# Filter 3: AND operator (multiple conditions)
query_where3 = """
SELECT FirstName, LastName, Department, Salary
FROM Employees
WHERE Department = 'IT' AND Salary > 70000
"""
display_sql_result(query_where3, "3. WHERE Department = 'IT' AND Salary > 70000 (AND operator)")
print("\n" + "="*60 + "\n")

# Filter 4: OR operator
query_where4 = """
SELECT FirstName, LastName, Department
FROM Employees
WHERE Department = 'IT' OR Department = 'Finance'
"""
display_sql_result(query_where4, "4. WHERE Department = 'IT' OR Department = 'Finance' (OR operator)")
print("\n" + "="*60 + "\n")

# Filter 5: IN operator (check if value is in a list)
query_where5 = """
SELECT FirstName, LastName, Department
FROM Employees
WHERE Department IN ('HR', 'Sales')
"""
display_sql_result(query_where5, "5. WHERE Department IN ('HR', 'Sales') (IN operator)")
print("\n" + "="*60 + "\n")

# Filter 6: BETWEEN operator (range of values)
query_where6 = """
SELECT FirstName, LastName, Salary
FROM Employees
WHERE Salary BETWEEN 60000 AND 75000
"""
display_sql_result(query_where6, "6. WHERE Salary BETWEEN 60000 AND 75000 (BETWEEN operator)")
print("\n" + "="*60 + "\n")

# Filter 7: LIKE operator (pattern matching)
query_where7 = """
SELECT FirstName, LastName, Email
FROM Employees
WHERE FirstName LIKE 'A%'
"""
display_sql_result(query_where7, "7. WHERE FirstName LIKE 'A%' (Starts with 'A')")
print("\n" + "="*60 + "\n")

# Filter 8: IS NULL / IS NOT NULL
query_where8 = """
SELECT FirstName, LastName, Email
FROM Employees
WHERE Email IS NOT NULL
"""
display_sql_result(query_where8, "8. WHERE Email IS NOT NULL (Check for NULL values)")
print("\n" + "="*60 + "\n")
```

## Section 6: Sorting and Limiting Results

Sort and paginate query results using ORDER BY and LIMIT.

```python
# Sort 1: ORDER BY ascending (default)
query_sort1 = """
SELECT FirstName, LastName, Salary
FROM Employees
ORDER BY Salary
"""
display_sql_result(query_sort1, "1. ORDER BY Salary (Ascending - default)")
print("\n" + "="*60 + "\n")

# Sort 2: ORDER BY descending
query_sort2 = """
SELECT FirstName, LastName, Salary
FROM Employees
ORDER BY Salary DESC
"""
display_sql_result(query_sort2, "2. ORDER BY Salary DESC (Descending)")
print("\n" + "="*60 + "\n")

# Sort 3: ORDER BY multiple columns
query_sort3 = """
SELECT FirstName, LastName, Department, Salary
FROM Employees
ORDER BY Department ASC, Salary DESC
"""
display_sql_result(query_sort3, "3. ORDER BY Multiple Columns (Department ASC, then Salary DESC)")
print("\n" + "="*60 + "\n")

# Sort 4: TOP N records using LIMIT
query_sort4 = """
SELECT FirstName, LastName, Salary
FROM Employees
ORDER BY Salary DESC
LIMIT 3
"""
display_sql_result(query_sort4, "4. LIMIT 3 - Top 3 Highest Paid Employees")
print("\n" + "="*60 + "\n")

# Sort 5: LIMIT with OFFSET (pagination)
query_sort5 = """
SELECT FirstName, LastName, Salary
FROM Employees
ORDER BY Salary DESC
LIMIT 2 OFFSET 2
"""
display_sql_result(query_sort5, "5. LIMIT 2 OFFSET 2 - Skip first 2 rows, get next 2 (Pagination)")
print("\n" + "="*60 + "\n")

# Sort 6: Lowest salaries
query_sort6 = """
SELECT FirstName, LastName, Salary
FROM Employees
ORDER BY Salary ASC
LIMIT 3
"""
display_sql_result(query_sort6, "6. LIMIT 3 - Bottom 3 Lowest Paid Employees")
print("\n" + "="*60 + "\n")
```

## Section 7: String and Numeric Functions

Learn how to use built-in SQL functions for string and numeric operations.

```python
# String Function 1: UPPER() - Convert to uppercase
query_string1 = """
SELECT 
    FirstName,
    UPPER(FirstName) AS 'First Name Upper',
    UPPER(Department) AS 'Department Upper'
FROM Employees
LIMIT 4
"""
display_sql_result(query_string1, "1. UPPER() - Convert strings to uppercase")
print("\n" + "="*60 + "\n")

# String Function 2: LOWER() - Convert to lowercase
query_string2 = """
SELECT 
    FirstName,
    LOWER(FirstName) AS 'First Name Lower',
    LOWER(Email) AS 'Email Lower'
FROM Employees
LIMIT 4
"""
display_sql_result(query_string2, "2. LOWER() - Convert strings to lowercase")
print("\n" + "="*60 + "\n")

# String Function 3: LENGTH() - Get string length
query_string3 = """
SELECT 
    FirstName,
    LENGTH(FirstName) AS 'Name Length',
    Email,
    LENGTH(Email) AS 'Email Length'
FROM Employees
LIMIT 4
"""
display_sql_result(query_string3, "3. LENGTH() - Get string length")
print("\n" + "="*60 + "\n")

# String Function 4: SUBSTR() - Get substring
query_string4 = """
SELECT 
    FirstName,
    SUBSTR(FirstName, 1, 3) AS 'First 3 Chars',
    Email,
    SUBSTR(Email, 1, 10) AS 'Email Start'
FROM Employees
LIMIT 4
"""
display_sql_result(query_string4, "4. SUBSTR() - Extract substring")
print("\n" + "="*60 + "\n")

# String Function 5: CONCAT() or || operator - Combine strings
query_string5 = """
SELECT 
    FirstName,
    LastName,
    FirstName || ' ' || LastName AS 'Full Name'
FROM Employees
LIMIT 4
"""
display_sql_result(query_string5, "5. Concatenation (||) - Combine strings")
print("\n" + "="*60 + "\n")

# Numeric Function 1: ABS() - Absolute value
# Let's create a calculation to show ABS
query_numeric1 = """
SELECT 
    FirstName,
    Salary,
    Salary - 70000 AS 'Difference from 70K',
    ABS(Salary - 70000) AS 'Absolute Difference'
FROM Employees
LIMIT 4
"""
display_sql_result(query_numeric1, "6. ABS() - Absolute value")
print("\n" + "="*60 + "\n")

# Numeric Function 2: ROUND() - Round numbers
query_numeric2 = """
SELECT 
    FirstName,
    Salary,
    Salary * 0.12 AS 'Exact Bonus (12%)',
    ROUND(Salary * 0.12, 2) AS 'Rounded Bonus'
FROM Employees
LIMIT 4
"""
display_sql_result(query_numeric2, "7. ROUND() - Round numbers to decimal places")
print("\n" + "="*60 + "\n")

# Numeric Function 3: MIN(), MAX(), AVG()
query_numeric3 = """
SELECT 
    MIN(Salary) AS 'Minimum Salary',
    MAX(Salary) AS 'Maximum Salary',
    ROUND(AVG(Salary), 2) AS 'Average Salary'
FROM Employees
"""
display_sql_result(query_numeric3, "8. MIN(), MAX(), AVG() - Aggregate functions")
print("\n" + "="*60 + "\n")
```

## Section 8: Aggregate Functions and GROUP BY

Learn how to group data and perform aggregate operations like SUM, AVG, COUNT.

```python
# GROUP BY 1: Basic group by with COUNT
query_group1 = """
SELECT 
    Department,
    COUNT(*) AS 'Number of Employees'
FROM Employees
GROUP BY Department
ORDER BY 'Number of Employees' DESC
"""
display_sql_result(query_group1, "1. GROUP BY Department - Count employees per department")
print("\n" + "="*60 + "\n")

# GROUP BY 2: GROUP BY with AVG
query_group2 = """
SELECT 
    Department,
    COUNT(*) AS 'Count',
    ROUND(AVG(Salary), 2) AS 'Average Salary',
    MIN(Salary) AS 'Min Salary',
    MAX(Salary) AS 'Max Salary'
FROM Employees
GROUP BY Department
ORDER BY 'Average Salary' DESC
"""
display_sql_result(query_group2, "2. GROUP BY with Multiple Aggregates (COUNT, AVG, MIN, MAX)")
print("\n" + "="*60 + "\n")

# GROUP BY 3: GROUP BY with SUM
query_group3 = """
SELECT 
    Department,
    COUNT(*) AS 'Number of Employees',
    SUM(Salary) AS 'Total Salary Budget'
FROM Employees
GROUP BY Department
ORDER BY 'Total Salary Budget' DESC
"""
display_sql_result(query_group3, "3. GROUP BY with SUM - Total salary per department")
print("\n" + "="*60 + "\n")

# GROUP BY 4: GROUP BY with condition in WHERE
query_group4 = """
SELECT 
    Department,
    COUNT(*) AS 'Active Employees',
    ROUND(AVG(Salary), 2) AS 'Average Salary'
FROM Employees
WHERE IsActive = 1
GROUP BY Department
"""
display_sql_result(query_group4, "4. GROUP BY with WHERE - Active employees only")
print("\n" + "="*60 + "\n")

# GROUP BY 5: Multiple columns in GROUP BY
query_group5 = """
SELECT 
    Department,
    IsActive,
    COUNT(*) AS 'Count'
FROM Employees
GROUP BY Department, IsActive
ORDER BY Department, IsActive
"""
display_sql_result(query_group5, "5. GROUP BY Multiple Columns (Department and IsActive)")
print("\n" + "="*60 + "\n")
```

## Section 9: HAVING Clause

Filter grouped results using HAVING (like WHERE but for aggregates).

```python
# HAVING 1: HAVING with COUNT condition
query_having1 = """
SELECT 
    Department,
    COUNT(*) AS 'Number of Employees'
FROM Employees
GROUP BY Department
HAVING COUNT(*) > 1
ORDER BY 'Number of Employees' DESC
"""
display_sql_result(query_having1, "1. HAVING COUNT(*) > 1 - Departments with more than 1 employee")
print("\n" + "="*60 + "\n")

# HAVING 2: HAVING with AVG condition
query_having2 = """
SELECT 
    Department,
    COUNT(*) AS 'Count',
    ROUND(AVG(Salary), 2) AS 'Average Salary'
FROM Employees
GROUP BY Department
HAVING AVG(Salary) > 65000
ORDER BY 'Average Salary' DESC
"""
display_sql_result(query_having2, "2. HAVING AVG(Salary) > 65000 - Departments with avg salary > 65K")
print("\n" + "="*60 + "\n")

# HAVING 3: HAVING with SUM condition
query_having3 = """
SELECT 
    Department,
    COUNT(*) AS 'Employees',
    SUM(Salary) AS 'Total Salary'
FROM Employees
GROUP BY Department
HAVING SUM(Salary) > 100000
ORDER BY 'Total Salary' DESC
"""
display_sql_result(query_having3, "3. HAVING SUM(Salary) > 100000 - Departments with total salary > 100K")
print("\n" + "="*60 + "\n")

# HAVING 4: WHERE and HAVING together
query_having4 = """
SELECT 
    Department,
    COUNT(*) AS 'Active Employees',
    ROUND(AVG(Salary), 2) AS 'Average Salary'
FROM Employees
WHERE IsActive = 1
GROUP BY Department
HAVING COUNT(*) >= 2
ORDER BY 'Active Employees' DESC
"""
display_sql_result(query_having4, "4. WHERE + HAVING - Active employees, departments with 2+ staff")
print("\n" + "="*60 + "\n")
```

## Section 10: JOIN Operations

Learn how to combine data from multiple tables using different types of JOINs.

```python
# JOIN 1: INNER JOIN - Only matching rows from both tables
query_join1 = """
SELECT 
    e.FirstName,
    e.LastName,
    e.Department,
    d.DepartmentName,
    d.Location
FROM Employees e
INNER JOIN Departments d ON e.Department = d.DepartmentName
ORDER BY e.Department
"""
display_sql_result(query_join1, "1. INNER JOIN - Employees with their department details")
print("\n" + "="*60 + "\n")

# JOIN 2: LEFT JOIN - All rows from left table, matching rows from right
query_join2 = """
SELECT 
    e.FirstName,
    e.LastName,
    e.Department,
    d.DepartmentName,
    d.Manager
FROM Employees e
LEFT JOIN Departments d ON e.Department = d.DepartmentName
ORDER BY e.Department
"""
display_sql_result(query_join2, "2. LEFT JOIN - All employees, with department info if available")
print("\n" + "="*60 + "\n")

# JOIN 3: Employees with their projects
query_join3 = """
SELECT 
    e.FirstName || ' ' || e.LastName AS 'Employee Name',
    e.Department,
    p.ProjectName,
    p.Budget,
    p.Status
FROM Employees e
INNER JOIN Projects p ON e.Department = p.Department
ORDER BY e.FirstName
"""
display_sql_result(query_join3, "3. Employees with Projects they're assigned to")
print("\n" + "="*60 + "\n")

# JOIN 4: Multiple JOINs
query_join4 = """
SELECT 
    e.FirstName,
    e.LastName,
    d.DepartmentName,
    d.Location,
    COUNT(p.ProjectID) AS 'Project Count'
FROM Employees e
INNER JOIN Departments d ON e.Department = d.DepartmentName
LEFT JOIN Projects p ON e.Department = p.Department
GROUP BY e.EmployeeID, d.DepartmentName, d.Location
ORDER BY e.FirstName
"""
display_sql_result(query_join4, "4. Multiple JOINs - Employees with departments and project counts")
print("\n" + "="*60 + "\n")

# JOIN 5: Join with aggregation
query_join5 = """
SELECT 
    d.DepartmentName,
    COUNT(e.EmployeeID) AS 'Employee Count',
    ROUND(AVG(e.Salary), 2) AS 'Avg Salary',
    COUNT(p.ProjectID) AS 'Project Count'
FROM Departments d
LEFT JOIN Employees e ON d.DepartmentName = e.Department
LEFT JOIN Projects p ON d.DepartmentName = p.Department
GROUP BY d.DepartmentName
"""
display_sql_result(query_join5, "5. JOINs with Aggregation - Department summary with projects")
print("\n" + "="*60 + "\n")
```

## Section 11: Subqueries and Complex Queries

Use subqueries to solve complex problems and create derived tables.

```python
# Subquery 1: Employees earning more than average
query_subquery1 = """
SELECT 
    FirstName,
    LastName,
    Salary
FROM Employees
WHERE Salary > (SELECT AVG(Salary) FROM Employees)
ORDER BY Salary DESC
"""
display_sql_result(query_subquery1, "1. SUBQUERY - Employees earning more than average salary")
print("\n" + "="*60 + "\n")

# Subquery 2: Employees in departments with high budgets
query_subquery2 = """
SELECT 
    e.FirstName,
    e.LastName,
    e.Department,
    e.Salary
FROM Employees e
WHERE e.Department IN (
    SELECT Department 
    FROM Projects 
    WHERE Budget > 50000
)
ORDER BY e.Department
"""
display_sql_result(query_subquery2, "2. SUBQUERY with IN - Employees in high-budget projects")
print("\n" + "="*60 + "\n")

# Subquery 3: Department stats with subquery
query_subquery3 = """
SELECT 
    Department,
    (SELECT COUNT(*) FROM Employees WHERE Department = e.Department) AS 'Employee Count',
    (SELECT AVG(Salary) FROM Employees WHERE Department = e.Department) AS 'Avg Salary',
    (SELECT SUM(Salary) FROM Employees WHERE Department = e.Department) AS 'Total Salary'
FROM Employees e
GROUP BY Department
"""
display_sql_result(query_subquery3, "3. SUBQUERY - Department statistics with correlated subqueries")
print("\n" + "="*60 + "\n")

# Subquery 4: Derived table
query_subquery4 = """
SELECT 
    dept_stats.Department,
    dept_stats.Avg_Salary,
    dept_stats.Employee_Count
FROM (
    SELECT 
        Department,
        ROUND(AVG(Salary), 2) AS Avg_Salary,
        COUNT(*) AS Employee_Count
    FROM Employees
    GROUP BY Department
) AS dept_stats
WHERE dept_stats.Avg_Salary > 60000
ORDER BY dept_stats.Avg_Salary DESC
"""
display_sql_result(query_subquery4, "4. DERIVED TABLE - Using subquery as a table")
print("\n" + "="*60 + "\n")

# Subquery 5: EXISTS clause
query_subquery5 = """
SELECT 
    d.DepartmentName
FROM Departments d
WHERE EXISTS (
    SELECT 1 FROM Employees e 
    WHERE e.Department = d.DepartmentName 
    AND e.Salary > 70000
)
"""
display_sql_result(query_subquery5, "5. EXISTS SUBQUERY - Departments with employees earning > 70K")
print("\n" + "="*60 + "\n")
```

## Section 12: UPDATE and DELETE Operations

Learn how to modify and delete data from tables.

```python
# Show current data before updates
print("BEFORE UPDATE:")
query_before = "SELECT * FROM Employees WHERE FirstName = 'Frank' LIMIT 1"
display_sql_result(query_before, "Current data for Frank")
print("\n" + "="*60 + "\n")

# UPDATE 1: Update single column
print("1. UPDATE - Change single column")
update1 = "UPDATE Employees SET Salary = 65000 WHERE FirstName = 'Frank'"
cursor.execute(update1)
conn.commit()
print("✓ Frank's salary updated to 65000")
print("\nAFTER UPDATE:")
display_sql_result(query_before, "")
print("\n" + "="*60 + "\n")

# UPDATE 2: Update multiple columns
print("2. UPDATE - Change multiple columns")
update2 = "UPDATE Employees SET Department = 'Marketing', IsActive = 1 WHERE FirstName = 'Frank'"
cursor.execute(update2)
conn.commit()
print("✓ Frank's department and status updated")
print("\nAFTER UPDATE:")
display_sql_result(query_before, "")
print("\n" + "="*60 + "\n")

# UPDATE 3: Update with condition
print("3. UPDATE with Condition - Give 5% raise to all IT employees")
update3 = "UPDATE Employees SET Salary = Salary * 1.05 WHERE Department = 'IT'"
cursor.execute(update3)
conn.commit()
print("✓ IT employees received 5% salary increase")
print("\nIT Employees after raise:")
query_it = "SELECT FirstName, LastName, Department, Salary FROM Employees WHERE Department = 'IT'"
display_sql_result(query_it, "")
print("\n" + "="*60 + "\n")

# DELETE 1: Delete specific row
print("4. DELETE - Remove employee")
print("\nDeletion note: We'll delete Frank's record as an example")
delete1 = "DELETE FROM Employees WHERE FirstName = 'Frank'"
cursor.execute(delete1)
conn.commit()
print("✓ Frank's record deleted from database")
print(f"✓ Total employees remaining: {cursor.execute('SELECT COUNT(*) FROM Employees').fetchone()[0]}")
print("\n" + "="*60 + "\n")

# Show sample of remaining employees
print("Remaining employees after deletion:")
query_all = "SELECT FirstName, LastName, Department FROM Employees LIMIT 5"
display_sql_result(query_all, "")
print("\n" + "="*60 + "\n")

# Note about DELETE without WHERE clause
print("⚠️  IMPORTANT WARNING:")
print("   DELETE FROM Employees;  ---> Would DELETE ALL employees!")
print("   Always use WHERE clause to specify which records to delete.")
print("\n" + "="*60 + "\n")
```

```python
print("=" * 60)
print("SQL DATA TYPES AND CONSTRAINTS REFERENCE")
print("=" * 60)

print("\n📌 PRIMARY DATA TYPES IN SQL:\n")

data_types_info = """
1. INTEGER / INT
   - Whole numbers (no decimal places)
   - Example: 100, -50, 0
   - Storage: 4 bytes (32-bit)

2. BIGINT
   - Large whole numbers
   - Storage: 8 bytes (64-bit)

3. FLOAT / REAL
   - Decimal numbers with limited precision
   - Example: 3.14, -0.5

4. DECIMAL / NUMERIC
   - Precise decimal numbers with specified precision
   - Syntax: DECIMAL(total_digits, decimal_places)
   - Example: DECIMAL(10, 2) → 12345678.90

5. TEXT / VARCHAR / CHAR
   - String data
   - VARCHAR(n): Variable length up to n characters
   - CHAR(n): Fixed length n characters
   - TEXT: Unlimited length

6. DATE
   - Calendar dates
   - Format: YYYY-MM-DD
   - Example: 2024-03-31

7. TIME
   - Time values
   - Format: HH:MM:SS
   - Example: 14:30:00

8. TIMESTAMP / DATETIME
   - Date and time combined
   - Example: 2024-03-31 14:30:00

9. BOOLEAN
   - True/False values (stored as 0 or 1)
"""

print(data_types_info)

print("\n" + "="*60)
print("CONSTRAINTS - Rules for data integrity:\n")

constraints_info = """
1. PRIMARY KEY
   - Uniquely identifies each row
   - Cannot be NULL
   - Only one per table
   - Example: EmployeeID INTEGER PRIMARY KEY

2. UNIQUE
   - All values must be unique
   - Can be NULL (unless NOT NULL is also specified)
   - Multiple UNIQUE constraints allowed
   - Example: Email TEXT UNIQUE

3. NOT NULL
   - Column must always have a value
   - Cannot be empty
   - Example: FirstName TEXT NOT NULL

4. DEFAULT
   - Provides default value if none specified
   - Example: IsActive BOOLEAN DEFAULT 1
   - If you don't specify, this value is used

5. CHECK
   - Validates data meets a condition
   - Example: CHECK(Age >= 18)
   - Salary DECIMAL CHECK(Salary > 0)

6. FOREIGN KEY
   - Links to PRIMARY KEY in another table
   - Ensures referential integrity
   - Example: DepartmentID INT REFERENCES Departments(ID)

7. AUTO_INCREMENT / AUTOINCREMENT
   - Automatically generates next number
   - Usually used with PRIMARY KEY
   - Example: ID INTEGER PRIMARY KEY AUTOINCREMENT
"""

print(constraints_info)

# Create a table with various constraints and data types
print("\n" + "="*60)
print("EXAMPLE: Table with Multiple Constraints\n")

create_products = """
CREATE TABLE IF NOT EXISTS Products (
    ProductID INTEGER PRIMARY KEY AUTOINCREMENT,
    ProductName TEXT NOT NULL,
    Category TEXT NOT NULL,
    Price DECIMAL(10, 2) NOT NULL CHECK(Price > 0),
    Stock INTEGER DEFAULT 0 CHECK(Stock >= 0),
    CreatedDate DATE DEFAULT CURRENT_DATE,
    IsActive BOOLEAN DEFAULT 1
)
"""

print("Table Definition:")
print(create_products)

cursor.execute(create_products)
conn.commit()

# Insert sample data
products_insert = """
INSERT INTO Products (ProductName, Category, Price, Stock) VALUES
('Laptop', 'Electronics', 999.99, 15),
('Mouse', 'Accessories', 29.99, 100),
('USB Cable', 'Accessories', 12.99, 500),
('Monitor', 'Electronics', 349.99, 8)
"""

cursor.execute(products_insert)
conn.commit()

print("\n✓ Products table created with constraints")
print("\nSample data in Products table:")
query_products = "SELECT * FROM Products"
display_sql_result(query_products, "")

print("\n" + "="*60)
print("✓ CONGRATULATIONS!")
print("=" * 60)
print("\nYou've completed the comprehensive SQL tutorial!")
print("\nTopics covered:")
print("  ✓ Database setup and table creation")
print("  ✓ Inserting data (single and multiple rows)")
print("  ✓ SELECT queries and filtering")
print("  ✓ Sorting and limiting results")
print("  ✓ String and numeric functions")
print("  ✓ Aggregation and GROUP BY")
print("  ✓ HAVING clauses")
print("  ✓ Different types of JOINs")
print("  ✓ Subqueries and complex queries")
print("  ✓ Updating and deleting data")
print("  ✓ Data types and constraints")
print("\nYou're now ready to use SQL for data management!")
print("=" * 60)
```
