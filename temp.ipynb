```python
import numpy
```

```python
matrix = [
    ["1","2",".",".","3",".",".",".","."],
    ["4",".",".","5",".",".",".",".","."],
    [".","9","8",".",".",".",".",".","3"],
    ["5",".",".",".","6",".",".",".","4"],
    [".",".",".","8",".","3",".",".","5"],
    ["7",".",".",".","2",".",".",".","6"],
    [".",".",".",".",".",".","2",".","."],
    [".",".",".","4","1","9",".",".","8"],
    [".",".",".",".","8",".",".","7","9"]
    ]
```

```python
ROWS= 9
COLUMNS = 9

ROW_VAL = {}
COL_VAL = {}
square = {}
# print("ROW VALUE")
for r in range(ROWS):
    for c in range(COLUMNS):

        value = matrix[r][c]
        if value == ".":
            continue
        
        ROW_VAL[r] = ROW_VAL.get(r,set())
        ROW_VAL[r].add(value) 
    

        # print("COLUMNS VALUE")

        COL_VAL[c] = COL_VAL.get(c,set())
        COL_VAL[c].add(value)

            # break

        
        # print("Square VALUE")
        square[(r//3, c//3)] = square.get((r//3, c//3), set())
        square[(r//3, c//3)].add(value)

```
