## Section 1: Import Required Libraries

```python
# Import required libraries
import pandas as pd                 # Pandas - for data manipulation and analysis
import numpy as np                  # NumPy - for numerical computations and arrays
from datetime import datetime, timedelta  # For handling date and time data
import random                       # For generating random numbers

# Set display options to see more columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

print("✓ All libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## Section 2: Create Dummy DataFrames

There are several ways to create a DataFrame in Pandas. Below are the most common methods:

```python
# Method 1: Create DataFrame from a Dictionary
# Dictionary keys become column names, values become the data
data_dict = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [25, 30, 35, 28, 32],
    'Salary': [50000, 60000, 75000, 55000, 80000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR']
}

df_dict = pd.DataFrame(data_dict)
print("DataFrame created from Dictionary:")
print(df_dict)
print("\n" + "="*60 + "\n")
```

```python
# Method 2: Create DataFrame from a List of Lists
# Each inner list becomes a row
data_list = [
    ['John', 45, 90000, 'Sales'],
    ['Jane', 38, 85000, 'Marketing'],
    ['Jack', 52, 95000, 'Finance'],
]

# Specify column names explicitly
df_list = pd.DataFrame(data_list, columns=['Name', 'Age', 'Salary', 'Department'])
print("DataFrame created from List of Lists:")
print(df_list)
print("\n" + "="*60 + "\n")
```

```python
# Method 3: Create DataFrame with Random Data using NumPy
# Useful for generating large datasets for testing and learning
np.random.seed(42)  # Set seed for reproducibility

df_random = pd.DataFrame({
    'StudentID': range(1, 11),  # IDs from 1 to 10
    'Score': np.random.randint(60, 100, 10),  # Random scores between 60-100
    'Attendance': np.random.uniform(0.5, 1.0, 10).round(2),  # Random values 0.5-1.0
    'Grade': np.random.choice(['A', 'B', 'C', 'D'], 10)  # Random grades
})

print("DataFrame created with Random Data:")
print(df_random)
print("\n" + "="*60 + "\n")
```

## Section 3: Explore DataFrame Structure

Understanding the structure and properties of your data is the first step in any data analysis task.

```python
# Use our main DataFrame (df_dict) for exploration
print("1. SHAPE - Returns dimensions (rows, columns)")
print(f"Shape: {df_dict.shape}")  # (5 rows, 4 columns)
print(f"Rows: {df_dict.shape[0]}, Columns: {df_dict.shape[1]}")
print("\n" + "-"*60 + "\n")

print("2. INFO() - Shows data types and memory usage")
df_dict.info()
print("\n" + "-"*60 + "\n")

print("3. DTYPES - Shows data type of each column")
print(df_dict.dtypes)
print("\n" + "-"*60 + "\n")

print("4. HEAD() - Shows first 5 rows (default)")
print(df_dict.head())
print("\n" + "-"*60 + "\n")

print("5. TAIL() - Shows last 3 rows")
print(df_dict.tail(3))
print("\n" + "-"*60 + "\n")

print("6. DESCRIBE() - Statistical summary of numeric columns")
print(df_dict.describe())
print("\n" + "="*60 + "\n")
```

## Section 4: Access DataFrame Data

There are multiple ways to access data from a DataFrame - by column, by row, or by specific cells.

```python
# Method 1: Access Single Column (returns Series)
print("1. Access Single Column - Using bracket notation")
print(df_dict['Name'])
print(f"Type: {type(df_dict['Name'])}")
print("\n" + "-"*60 + "\n")

# Method 2: Access Multiple Columns (returns DataFrame)
print("2. Access Multiple Columns")
print(df_dict[['Name', 'Salary']])
print(f"Type: {type(df_dict[['Name', 'Salary']])}")
print("\n" + "-"*60 + "\n")

# Method 3: Access by Row Position using iloc[] (index-location)
print("3. Access row by position using iloc[] (row 1, index 1)")
print(df_dict.iloc[1])  # 0-indexed, so index 1 is the 2nd row
print("\n" + "-"*60 + "\n")

# Method 4: Access specific cell using iloc[]
print("4. Access specific cell (row 0, column 1)")
print(f"Value: {df_dict.iloc[0, 1]}")  # First row, second column
print("\n" + "-"*60 + "\n")

# Method 5: Access by row label using loc[] (label-location)
print("5. Access row by index label using loc[]")
print(df_dict.loc[0])  # Using index label 0
print("\n" + "-"*60 + "\n")

# Method 6: Access specific value using loc[] with label
print("6. Access specific value using loc[] (row 0, column 'Age')")
print(f"Value: {df_dict.loc[0, 'Age']}")
print("\n" + "="*60 + "\n")
```

## Section 5: Data Filtering and Selection

Filtering is one of the most common operations in data analysis. Learn different ways to filter your data.

```python
# Display the original dataframe first
print("Original DataFrame:")
print(df_dict)
print("\n" + "="*60 + "\n")

# Filter 1: Get rows where Age > 28
print("Filter 1: Employees with Age > 28")
age_filter = df_dict[df_dict['Age'] > 28]
print(age_filter)
print("\n" + "-"*60 + "\n")

# Filter 2: Get rows where Department == 'IT'
print("Filter 2: Employees in IT department")
dept_filter = df_dict[df_dict['Department'] == 'IT']
print(dept_filter)
print("\n" + "-"*60 + "\n")

# Filter 3: Combine multiple conditions using & (AND) and | (OR)
print("Filter 3: Employees with Age > 28 AND Salary > 60000")
combined_filter = df_dict[(df_dict['Age'] > 28) & (df_dict['Salary'] > 60000)]
print(combined_filter)
print("\n" + "-"*60 + "\n")

# Filter 4: Using .isin() to filter by multiple values
print("Filter 4: Employees in HR or Finance department")
dept_list = ['HR', 'Finance']
dept_filter2 = df_dict[df_dict['Department'].isin(dept_list)]
print(dept_filter2)
print("\n" + "-"*60 + "\n")

# Filter 5: Using .query() method (SQL-like syntax)
print("Filter 5: Using query() - Employees with Salary >= 60000")
query_result = df_dict.query('Salary >= 60000')
print(query_result)
print("\n" + "="*60 + "\n")
```

## Section 6: Basic DataFrame Operations

Learn how to sort, group, and aggregate data to extract meaningful insights.

```python
# Operation 1: Sorting - sort by Age in ascending order
print("1. SORT - Sort by Age (ascending)")
sorted_age = df_dict.sort_values('Age')
print(sorted_age)
print("\n" + "-"*60 + "\n")

# Operation 2: Sort by Salary in descending order
print("2. SORT - Sort by Salary (descending)")
sorted_salary = df_dict.sort_values('Salary', ascending=False)
print(sorted_salary)
print("\n" + "-"*60 + "\n")

# Operation 3: Grouping - Count employees by department
print("3. GROUP BY - Count employees by Department")
dept_count = df_dict.groupby('Department').size()
print(dept_count)
print("\n" + "-"*60 + "\n")

# Operation 4: Grouping - Get average salary by department
print("4. GROUP BY - Average Salary by Department")
avg_salary = df_dict.groupby('Department')['Salary'].mean()
print(avg_salary)
print("\n" + "-"*60 + "\n")

# Operation 5: Multiple aggregations
print("5. GROUP BY - Multiple aggregations on Age group")
grouped_stats = df_dict.groupby('Department').agg({
    'Age': 'mean',           # Average age
    'Salary': ['min', 'max', 'mean'],  # Min, max, and mean salary
    'Name': 'count'          # Count of employees
})
print(grouped_stats)
print("\n" + "-"*60 + "\n")

# Operation 6: Mathematical operations on columns
print("6. Mathematical Operations - Create new column with bonus (10% of salary)")
df_dict['Bonus'] = df_dict['Salary'] * 0.10
print(df_dict)
print("\n" + "="*60 + "\n")
```

## Section 7: Data Cleaning Basics

Real-world data often contains missing values, duplicates, and inconsistencies. Learn how to handle them.

```python
# Create a DataFrame with missing values and duplicates
data_messy = {
    'Product': ['A', 'B', 'C', 'A', 'B', None, 'D'],
    'Price': [10, 20, None, 10, 25, 30, 35],
    'Quantity': [5, 10, 15, 5, None, 25, 30]
}
df_messy = pd.DataFrame(data_messy)

print("Original DataFrame with Missing Values:")
print(df_messy)
print("\n" + "-"*60 + "\n")

# Cleaning 1: Check for missing values
print("1. Check for Missing Values using isnull() or isna()")
print(df_messy.isnull())
print("\n" + "-"*60 + "\n")

# Cleaning 2: Count missing values
print("2. Count Missing Values in Each Column")
print(df_messy.isnull().sum())
print("\n" + "-"*60 + "\n")

# Cleaning 3: Fill missing values with a specific value
print("3. Fill Missing Values with 0")
df_filled = df_messy.fillna(0)
print(df_filled)
print("\n" + "-"*60 + "\n")

# Cleaning 4: Fill missing values with forward fill (propagate last valid value)
print("4. Fill Missing Values with Forward Fill")
df_ffill = df_messy.ffill()  # Forward fill - propagate last valid value
print(df_ffill)
print("\n" + "-"*60 + "\n")

# Cleaning 5: Drop rows with missing values
print("5. Drop Rows with Any Missing Values")
df_dropped = df_messy.dropna()
print(df_dropped)
print("\n" + "-"*60 + "\n")

# Cleaning 6: Handle duplicates
print("6. Check for Duplicate Rows")
print(df_messy.duplicated())
print("\n" + "-"*60 + "\n")

# Cleaning 7: Remove duplicates
print("7. Remove Duplicate Rows")
df_no_dupes = df_messy.drop_duplicates()
print(df_no_dupes)
print("\n" + "-"*60 + "\n")

# Cleaning 8: Change data types
print("8. Change Data Types using astype()")
df_messy['Quantity'] = df_messy['Quantity'].astype('Int64')  # Use nullable integer type
print(df_messy.dtypes)
print("\n" + "="*60 + "\n")
```

## Section 8: Working with Rows and Columns

Learn how to add, remove, rename, and modify rows and columns in your DataFrame.

```python
# Create a copy of df_dict to work with
df_work = df_dict.copy()

print("Original DataFrame:")
print(df_work)
print("\n" + "="*60 + "\n")

# Operation 1: Add a new column
print("1. Add a New Column - Years of Experience")
df_work['Experience'] = [3, 8, 12, 5, 10]
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 2: Rename columns
print("2. Rename Columns")
df_work = df_work.rename(columns={
    'Name': 'Employee_Name',
    'Age': 'Employee_Age',
    'Department': 'Dept'
})
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 3: Drop a column
print("3. Drop a Column")
df_work = df_work.drop('Experience', axis=1)  # axis=1 means column, axis=0 means row
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 4: Reorder columns
print("4. Reorder Columns")
new_order = ['Employee_Name', 'Dept', 'Salary', 'Employee_Age', 'Bonus']
df_work = df_work[new_order]
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 5: Add a new row
print("5. Add a New Row using concat()")
new_row = pd.DataFrame({
    'Employee_Name': ['Frank'],
    'Dept': ['Marketing'],
    'Salary': [70000],
    'Employee_Age': [35],
    'Bonus': [7000]
})
df_work = pd.concat([df_work, new_row], ignore_index=True)
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 6: Drop a row
print("6. Drop Row by Index")
df_work = df_work.drop(0)  # Drop the first row
print(df_work)
print("\n" + "-"*60 + "\n")

# Operation 7: Reset index after dropping rows
print("7. Reset Index After Dropping Rows")
df_work = df_work.reset_index(drop=True)  # drop=True removes old index as a column
print(df_work)
print("\n" + "="*60 + "\n")

print("✓ All operations completed successfully!")
print("You now have a solid foundation in Pandas fundamentals!")
```

## Section 9: Data Type Conversion and Type Changes

Learn how to convert and change data types of columns in your DataFrame.

```python
# Create a sample DataFrame with different data types
data_types = {
    'ID': ['1', '2', '3', '4', '5'],  # String numbers
    'Price': [10.5, 20.0, 15.75, 30.0, 25.5],  # Float
    'Quantity': ['5', '10', '15', '20', '25'],  # String numbers
    'Available': ['Yes', 'No', 'Yes', 'Yes', 'No'],  # String boolean
    'Created_Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05']  # String dates
}
df_types = pd.DataFrame(data_types)

print("Original DataFrame with Mixed Data Types:")
print(df_types)
print("\nData Types:")
print(df_types.dtypes)
print("\n" + "="*60 + "\n")

# Type Conversion 1: Convert string to integer using astype()
print("1. Convert String to Integer (ID column)")
df_types['ID'] = df_types['ID'].astype('int')
print(df_types['ID'])
print(f"Data type: {df_types['ID'].dtype}")
print("\n" + "-"*60 + "\n")

# Type Conversion 2: Convert string to integer (Quantity)
print("2. Convert String to Integer (Quantity column)")
df_types['Quantity'] = df_types['Quantity'].astype('int')
print(df_types['Quantity'])
print("\n" + "-"*60 + "\n")

# Type Conversion 3: Convert float to integer
print("3. Convert Float to Integer (Price column)")
df_types['Price_Int'] = df_types['Price'].astype('int')  # Truncates decimals
print(df_types['Price_Int'])
print("\n" + "-"*60 + "\n")

# Type Conversion 4: Convert string to datetime
print("4. Convert String to Datetime")
df_types['Created_Date'] = pd.to_datetime(df_types['Created_Date'])
print(df_types['Created_Date'])
print(f"Data type: {df_types['Created_Date'].dtype}")
print("\n" + "-"*60 + "\n")

# Type Conversion 5: Convert to categorical (for memory efficiency)
print("5. Convert String to Categorical (Available column)")
df_types['Available'] = df_types['Available'].astype('category')
print(df_types['Available'])
print(f"Data type: {df_types['Available'].dtype}")
print(f"Categories: {df_types['Available'].cat.categories.tolist()}")
print("\n" + "-"*60 + "\n")

# Type Conversion 6: Convert boolean values
print("6. Convert String to Boolean")
df_types['Available_Bool'] = (df_types['Available'] == 'Yes')
print(df_types['Available_Bool'])
print(f"Data type: {df_types['Available_Bool'].dtype}")
print("\n" + "-"*60 + "\n")

# Type Conversion 7: Check data types and convert all at once
print("7. Convert Multiple Columns at Once")
df_types = df_types.astype({
    'ID': 'int',
    'Quantity': 'int',
    'Price': 'float32'  # Use float32 for memory efficiency
})
print(df_types.dtypes)
print("\n" + "="*60 + "\n")
```

## Section 10: The apply() Function

The apply() function allows you to apply custom functions to DataFrame columns or rows. It's very powerful for complex transformations.

```python
# Create a sample DataFrame for apply() examples
df_apply = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
    'Score': [85, 92, 78, 88],
    'Grade': ['B', 'A', 'C', 'B'],
    'Age': [22, 25, 20, 23]
})

print("Original DataFrame:")
print(df_apply)
print("\n" + "="*60 + "\n")

# Apply Example 1: Apply lambda function to single column
print("1. Apply Lambda Function to Score Column")
# Convert scores to percentage (multiply by different factor)
df_apply['Score_Adjusted'] = df_apply['Score'].apply(lambda x: x * 1.1)
print(df_apply[['Name', 'Score', 'Score_Adjusted']])
print("\n" + "-"*60 + "\n")

# Apply Example 2: Apply custom function
print("2. Apply Custom Function - Assign Grade Based on Score")

def assign_grade(score):
    """Custom function to assign letter grades based on score"""
    if score >= 90:
        return 'A'
    elif score >= 80:
        return 'B'
    elif score >= 70:
        return 'C'
    else:
        return 'D'

# Note: We're not modifying df_apply['Grade'], just showing the function
grades = df_apply['Score'].apply(assign_grade)
print(grades)
print("\n" + "-"*60 + "\n")

# Apply Example 3: Apply function to multiple columns (axis=1 means row-wise)
print("3. Apply Function Row-wise (axis=1)")

def calculate_performance(row):
    """Function that uses multiple columns from a row"""
    score = row['Score']
    age = row['Age']
    # Calculate performance ratio
    return f"{row['Name']}: Score={score}, Age={age}, Ratio={score/age:.2f}"

performance = df_apply.apply(calculate_performance, axis=1)
print(performance)
print("\n" + "-"*60 + "\n")

# Apply Example 4: String operations with apply
print("4. Apply String Operations")
df_apply['Name_Length'] = df_apply['Name'].apply(len)
df_apply['Name_Upper'] = df_apply['Name'].apply(str.upper)
print(df_apply[['Name', 'Name_Length', 'Name_Upper']])
print("\n" + "-"*60 + "\n")

# Apply Example 5: Apply with conditional logic
print("5. Apply Conditional Function")

def pass_fail(score):
    """Simple pass/fail logic - lambda alternative"""
    return 'Pass' if score >= 70 else 'Fail'

df_apply['Status'] = df_apply['Score'].apply(pass_fail)
print(df_apply[['Name', 'Score', 'Status']])
print("\n" + "-"*60 + "\n")

# Apply Example 6: Apply map() for value replacement (Series only)
print("6. Apply map() Function for Value Replacement")
grade_description = {
    'A': 'Excellent',
    'B': 'Good',
    'C': 'Average',
    'D': 'Poor'
}
df_apply['Grade_Description'] = df_apply['Grade'].map(grade_description)
print(df_apply[['Name', 'Grade', 'Grade_Description']])
print("\n" + "="*60 + "\n")
```

## Section 11: Useful Pandas Functions

Explore important pandas functions like value_counts(), unique(), nunique(), and more.

```python
# Create sample data for useful functions
data_products = {
    'Product': ['Laptop', 'Mouse', 'Keyboard', 'Monitor', 'Mouse', 'Laptop', 'USB Cable', 'Monitor'],
    'Category': ['Electronics', 'Accessories', 'Accessories', 'Electronics', 'Accessories', 'Electronics', 'Accessories', 'Electronics'],
    'Price': [1200, 25, 75, 350, 25, 1200, 10, 350],
    'Stock': [5, 50, 30, 10, 50, 5, 100, 10]
}
df_products = pd.DataFrame(data_products)

print("Sample Product DataFrame:")
print(df_products)
print("\n" + "="*60 + "\n")

# Function 1: value_counts() - Count occurrences of each unique value
print("1. value_counts() - Count Each Product")
product_counts = df_products['Product'].value_counts()
print(product_counts)
print("\n" + "-"*60 + "\n")

# Function 2: unique() - Get unique values
print("2. unique() - Get Unique Products")
unique_products = df_products['Product'].unique()
print(unique_products)
print(f"Type: {type(unique_products)}")
print("\n" + "-"*60 + "\n")

# Function 3: nunique() - Count number of unique values
print("3. nunique() - Count Unique Products")
num_unique = df_products['Product'].nunique()
print(f"Number of unique products: {num_unique}")
print("\n" + "-"*60 + "\n")

# Function 4: duplicated() - Find duplicate rows
print("4. duplicated() - Find Duplicate Rows")
duplicates = df_products.duplicated()
print(duplicates)
print("\nDuplicate rows:")
print(df_products[df_products.duplicated(keep=False)].sort_values('Product'))
print("\n" + "-"*60 + "\n")

# Function 5: isin() - Check if values exist in a list
print("5. isin() - Find Electronics Category")
electronics = df_products[df_products['Category'].isin(['Electronics'])]
print(electronics)
print("\n" + "-"*60 + "\n")

# Function 6: str functions on string columns
print("6. String Operations - Contains 'o' in Product Name")
has_o = df_products['Product'].str.contains('o', case=False)
print(df_products[has_o])
print("\n" + "-"*60 + "\n")

# Function 7: between() - Filter values within range
print("7. between() - Products Priced Between 20 and 100")
mid_price = df_products[df_products['Price'].between(20, 100)]
print(mid_price)
print("\n" + "-"*60 + "\n")

# Function 8: nlargest() and nsmallest()
print("8. nlargest() - Top 3 Most Expensive Products")
top_expensive = df_products.nlargest(3, 'Price')[['Product', 'Price']]
print(top_expensive)
print("\n" + "-"*60 + "\n")

# Function 9: idxmax() and idxmin() - Get index of max/min values
print("9. idxmax() and idxmin() - Index of Max and Min Price")
max_price_idx = df_products['Price'].idxmax()
min_price_idx = df_products['Price'].idxmin()
print(f"Most expensive product (index {max_price_idx}): {df_products.loc[max_price_idx, 'Product']} - ${df_products.loc[max_price_idx, 'Price']}")
print(f"Cheapest product (index {min_price_idx}): {df_products.loc[min_price_idx, 'Product']} - ${df_products.loc[min_price_idx, 'Price']}")
print("\n" + "-"*60 + "\n")

# Function 10: value_counts() with normalize
print("10. value_counts(normalize=True) - Proportion of Each Category")
category_proportion = df_products['Category'].value_counts(normalize=True)
print(category_proportion)
print("\n" + "="*60 + "\n")
```

## Section 12: Merging, Joining, and Combining DataFrames

Learn how to combine multiple DataFrames using merge(), join(), and concat() operations.

```python
# Create sample DataFrames for merging examples
employees = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR']
})

salaries = pd.DataFrame({
    'EmployeeID': [1, 2, 3, 4, 5],
    'Salary': [50000, 70000, 75000, 68000, 55000],
    'Bonus': [5000, 7000, 7500, 6800, 5500]
})

projects = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'ProjectName': ['Project A', 'Project B', 'Project C'],
    'Hours': [40, 35, 50]
})

print("Employees DataFrame:")
print(employees)
print("\n" + "-"*60 + "\n")

print("Salaries DataFrame:")
print(salaries)
print("\n" + "-"*60 + "\n")

print("Projects DataFrame:")
print(projects)
print("\n" + "="*60 + "\n")

# Merge 1: Inner merge (default)
print("1. INNER MERGE - Only matching records in both DataFrames")
merged_inner = pd.merge(employees, salaries, on='EmployeeID')
print(merged_inner)
print("\n" + "-"*60 + "\n")

# Merge 2: Left merge
print("2. LEFT MERGE - All records from left DataFrame")
merged_left = pd.merge(employees, projects, on='EmployeeID', how='left')
print(merged_left)
print("\n" + "-"*60 + "\n")

# Merge 3: Right merge
print("3. RIGHT MERGE - All records from right DataFrame")
merged_right = pd.merge(employees, projects, on='EmployeeID', how='right')
print(merged_right)
print("\n" + "-"*60 + "\n")

# Merge 4: Full (outer) merge
print("4. OUTER MERGE - All records from both DataFrames")
merged_outer = pd.merge(employees, projects, on='EmployeeID', how='outer')
print(merged_outer)
print("\n" + "-"*60 + "\n")

# Merge 5: Merge multiple DataFrames
print("5. MERGE Multiple DataFrames")
# First merge employees and salaries
temp = pd.merge(employees, salaries, on='EmployeeID')
# Then merge with projects
final_merged = pd.merge(temp, projects, on='EmployeeID', how='left')
print(final_merged)
print("\n" + "-"*60 + "\n")

# Combine 6: Using concat() for row-wise combination
print("6. CONCAT - Combine DataFrames Row-wise")
df_q1 = pd.DataFrame({
    'Quarter': ['Q1', 'Q1', 'Q1'],
    'Sales': [10000, 15000, 12000]
})
df_q2 = pd.DataFrame({
    'Quarter': ['Q2', 'Q2', 'Q2'],
    'Sales': [18000, 22000, 19000]
})
combined_quarters = pd.concat([df_q1, df_q2], ignore_index=True)
print(combined_quarters)
print("\n" + "-"*60 + "\n")

# Combine 7: Using concat() for column-wise combination
print("7. CONCAT - Combine DataFrames Column-wise (axis=1)")
df_names = pd.DataFrame({'Name': ['Alice', 'Bob', 'Charlie']})
df_ages = pd.DataFrame({'Age': [25, 30, 35]})
combined_cols = pd.concat([df_names, df_ages], axis=1)
print(combined_cols)
print("\n" + "-"*60 + "\n")

# Combine 8: Merge on different column names
print("8. MERGE on Different Column Names")
df_a = pd.DataFrame({'ID': [1, 2, 3], 'Value_A': [10, 20, 30]})
df_b = pd.DataFrame({'Code': [1, 2, 3], 'Value_B': [100, 200, 300]})
merged_diff = pd.merge(df_a, df_b, left_on='ID', right_on='Code')
print(merged_diff)
print("\n" + "="*60 + "\n")

print("✓ Congratulations! You now have comprehensive knowledge of Pandas!")
print("You've learned:")
print("  ✓ DataFrame creation and exploration")
print("  ✓ Data access and filtering")
print("  ✓ Data cleaning and type conversion")
print("  ✓ apply() and map() functions")
print("  ✓ Useful Series/DataFrame functions")
print("  ✓ Merging and combining DataFrames")
```
